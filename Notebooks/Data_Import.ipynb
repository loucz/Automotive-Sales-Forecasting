## Fuel Prices Extraction

I imported the weekly prices of gasoline and diesel with and without taxes from an excel file downloaded from EU website.
Then I calculate the monthly average price for them.

```python
import pandas as pd
import datetime 
```

```python
fp=pd.read_excel('/users/loukas/Desktop/thesis/dataset/fuel_prices.xlsx',sheet_name=1,index_col='Date')
```

```python
fp
```

```python
fp.dtypes
```

```python
fp=fp.astype('str')
fp.dtypes
```

```python
gas=[]
dsl=[]

for i in fp['gas_wt']:
    gas.append(i.replace(',',''))
for i in fp['dsl_wt']:
    dsl.append(i.replace(',',''))
    
```

```python
fp['gas_wt']=gas
```

```python
fp['dsl_wt']=dsl
```

```python
fp=fp.astype('float')
fp.dtypes
```

```python
m=fp.index.month
y=fp.index.year
```

```python
m=m.astype('string')
```

```python
y=y.astype('string')
```

```python
fp['y-m']=y+'-'+m
```

```python
fp
```

```python
avg_gas_wt=[]
avg_dsl_wt=[]
avg_gas_wot=[]
avg_dsl_wot=[]



for i in fp['y-m'].unique():
    avg_gas_wt.append(fp.loc[fp['y-m']==i,'gas_wt'].mean())
    avg_dsl_wt.append(fp.loc[fp['y-m']==i,'dsl_wt'].mean())
    avg_gas_wot.append(fp.loc[fp['y-m']==i,'gas_wot'].mean())
    avg_dsl_wot.append(fp.loc[fp['y-m']==i,'dsl_wot'].mean())
```

```python
fp1=pd.DataFrame({'gas_wt':avg_gas_wt,'dsl_wt':avg_dsl_wt,'gas_wot':avg_gas_wot,'dsl_wot':avg_dsl_wot},index=fp['y-m'].unique())
```

```python
fp1
```

```python
fp1=fp1.iloc[::-1]
```

```python
fp1.to_excel('/Users/loukas/Desktop/thesis/dataset/fp_OUT.xls')
```

## Athens Stock Exchange General Index Extraction

From an excel file with the daily prices of the Athens Stock Exchabge General Index I extracted the monthly average and close price and I transformed the index column to month and year.

```python
import numpy as np
```

```python
df_gd=pd.read_csv('/Users/loukas/Desktop/thesis/dataset/HistoryClosesGDall.csv',header=1,names=['DATE','CLOSE','CHANGE%','OPEN','HIGH','LOW','VOLUME','TURNOVER'],index_col='DATE')
```

```python
df_gd
```

```python
df1_gd=df_gd.loc[:'05/1/2004',:]
```

```python
df1_gd
```

```python
df1_gd=df1_gd.iloc[::-1]
```

```python
df1_gd
```

```python
ym=[]

for i in df1_gd.index:
    if len(i)==9:
        m='0'+i[3]
        y=i[5:9]
    elif len(i)==10:
        m=i[3]+i[4]
        y=i[6:10]
    else:
        print("ERROR")
    
    ym.append(y+'-'+m)
```

```python
df1_gd=df1_gd.astype('string')
```

```python
df1_gd.dtypes
```

```python
df1_gd
```

```python
close=[]
change=[]

for i in df1_gd['CLOSE']:
    close.append(i.replace(',','.'))
for i in df1_gd['CHANGE%']:
    change.append(i.replace(',','.'))
```

```python
dd=pd.DataFrame({'CLOSE':close,'CHANGE%':change,'Y-M':ym}, index=df1_gd.index,dtype=float)
```

```python
dd
```

```python
dd.loc[dd['Y-M']=='2004-01','CLOSE']
```

```python
dd.loc[dd['Y-M']=='2004-01','CLOSE'].mean()
```

```python
dd.loc[dd['Y-M']=='2004-01','CLOSE'][-1]
```

```python
len(dd['Y-M'].unique())
```

```python
dd.dtypes
```

```python
avg=[]
last_day=[]

for i in dd['Y-M'].unique():
    avg.append(dd.loc[dd['Y-M']==i,'CLOSE'].mean())
    last_day.append(dd.loc[dd['Y-M']==i,'CLOSE'][-1])
```

```python
gd=pd.DataFrame({'average':avg,'last_day':last_day},index=dd['Y-M'].unique())
```

```python
gd.to_excel('/Users/loukas/Desktop/thesis/dataset/gd_atOUT.xls')
```
