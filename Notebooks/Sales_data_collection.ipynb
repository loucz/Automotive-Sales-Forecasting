# Import past monthly sales data

#### Every month an excel file is released by Hellenic Association Importer Representatives (https://seaa.gr). In every file there are the monthly sales of each car brand being sold in Greece. I downloaded all the excel files in a folder and with the code below I extracted from them all the monthly sales data for the car brands needed for the dissertation. Doing some string manipulation I built a DataFrame for the all the brands and export it to an excel file.

#### Concatenate all the monthly sales columns in one DataFrame

```python
all_data = pd.DataFrame()
for f in li:
    df = pd.read_excel(f,usecols=[1,2])
    df.insert(2,'file_name',f)
    all_data = all_data.append(df)
```

```python
all_data
```

```python
data=all_data.dropna()
```

```python
data=data.rename(columns={'Unnamed: 1':'brands','Unnamed: 2':'sales'})
```

```python
data
```

```python
data=data.reset_index(drop=True)
```

```python
data
```

```python
data.file_name.unique()
```

#### Retrieve only the necessary car brands and put them in DataFrame

```python
TOYOTA=[]
FORD=[]
OPEL=[]
AUDI=[]
FIAT=[]
date=[]

for ind in data.index:
    if data['brands'][ind]=='TOYOTA':
        TOYOTA.append(data['sales'][ind])
        date.append(data['file_name'][ind])
        
    elif data['brands'][ind]=='AUDI':
        AUDI.append(data['sales'][ind])
        
    elif data['brands'][ind]=='OPEL':
        OPEL.append(data['sales'][ind])
        
    elif data['brands'][ind]=='FORD':
        FORD.append(data['sales'][ind])
        
    elif data['brands'][ind]=='FIAT':
        FIAT.append(data['sales'][ind])
        
```

```python
df1=pd.DataFrame({'TOYOTA':TOYOTA,'AUDI':AUDI,'OPEL':OPEL,'FORD':FORD,'FIAT':FIAT,'file_date':date})
```

```python
df1.info()
```

```python
df1
```

#### String Manipulation to create Year-Month Index 

```python
date1=[]
for i in df1['file_date']:
    date1.append(i[43:50])
```

```python
date1
```

```python
date2=[]
for i in date1:
    if i[6]=="0" or i[6]=="1" or i[6]=="2":
        date2.append(i)
    else:
        date2.append(i[0:6])

date3=[]
for i in date2:
    if len(i)==6:
        date3.append(i[0:5]+ "0" +i[5:7])
    else:
        date3.append(i)
        
```

```python
date3
```

```python
df1.insert(5,'date',date3)
```

```python
df1
```

### Final DataFrame and export to excel 

```python
df2=df1.sort_values(by=['date'])
```

```python
df2.set_index('date')
```
